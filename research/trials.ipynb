# ensure_annotations 

```python
def get_product(x:int, y:int):
    return x*y
```

```python
get_product(x=2,y=4)
```

```python
get_product(x=2,y="4")
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(x:int, y:int):
    return x*y
```

```python
get_product(x=2,y=4)
```
